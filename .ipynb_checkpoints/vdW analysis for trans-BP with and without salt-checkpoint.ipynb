# I am going to define a function for this plot so that I can apply to all plots 

## function for the histogram

```python
def histplot(step):
    resi_name='resi-'+str(step)

    fig, ax = plt.subplots(figsize=(10,10))
    sns.distplot(df[resi_name],bins=100,color='blue',label="no salt-"+resi_name)
    sns.distplot(df_s[resi_name],bins=100,color='orange',label="salt-"+resi_name)
    plt.legend()
    
    sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 2.5})
    ax.set_xlabel("vdw in kcal/mol")
    plt.savefig("vdw"+str(step)+".png")
```

```python
histplot(430)
```

## function for the time dependence plot 

```python
def timeplot(step):
    resi_name='resi-'+str(step)
    fig, ax = plt.subplots(figsize=(10,5))
    plot(df[resi_name],color='blue',alpha=0.5, label="no salt-"+resi_name)
    plot(df_s[resi_name],color='orange',alpha=0.5, label= "salt-"+resi_name)
    plt.legend()
    ax.set_xlabel("Time (10 ps)")
    ax.set_ylabel("vdw (kcal/mol)")
   
    plt.savefig("time_dep_vdw"+str(step)+".png")
```

```python
timeplot(430)
```

## let's enumerate all steps

```python
for i in range(430,446):
    print i
    histplot(i)
    timeplot(i)
```

## Combine figures into one horizontal strip

```python
import Image
blank_img=Image.new("RGB",(16*720,720))
I=range(430,446)
J=range(0,16)
for i,j in zip(I,J):
    single_img=Image.open("vdw"+str(i)+".png")
    blank_img.paste(single_img,(j*720,0))
blank_img.save("vdw_hist_all.jpg",dpi=(16*720,720))
```
